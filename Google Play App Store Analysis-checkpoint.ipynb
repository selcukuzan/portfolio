# Separating Non-English Apps

```python
def is_english(string):
    non_ascii = 0
    
    for character in string:
        if ord(character) > 127:
            non_ascii += 1
    if non_ascii > 3:
        return False
    else:
        return True
    
english_apps = list()
non_english_apps = list()

for i in df["App"]:
    if is_english(i) is True:
        english_apps.append(i)
    else:
        non_english_apps.append(i)
        
english_apps
```
