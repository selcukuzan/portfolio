```python
import pandas as pd
import numpy as np
pd.options.display.max_rows = 99
```

```python
df = pd.read_csv("autos.csv" , encoding="Latin-1")
df.head(5)
```

```python
new_columns = []
for c in df.columns:
    modified_columns = c.strip()
    new_columns.append(modified_columns)
df.columns = new_columns
```

```python
df.head(5)
```

```python
cols = list(df.columns)
cols
```

```python
df = df.dropna()
df.notnull().sum()
```

```python
df.info()
```

```python
clean_cols = {"dateCrawled":"Date_Crawled","name":"Name","seller":"Seller","offerType":"Offer_Type","price":"Price","abtest":"AB_Test","vehicleType":"Vehicle_Type","yearOfRegistration":"Year_Of_Registration","gearbox":"Gear_Box","powerPS":"Power_PS","model":"Model","kilometer":"Kilometer","monthOfRegistration":"Month_Of_Registration","fuelType":"Fuel_Type","brand":"Brand","notRepairedDamage":"Not_Repaired_Damage","dateCreated":"Date_Created","nrOfPictures":"Nr_Of_Pictures","postalCode":"Postal_Code","lastSeen":"Last_Seen"}
clean_cols_2 = ["Date_Crawled","Name","Seller","Offer_Type","Price","AB_Test","Vehicle_Type","Year_Of_Registration","Gear_Box","Power_PS","Model","Kilometer","Month_Of_Registration","Fuel_Type","Brand","Not_Repaired_Damage","Date_Created","Nr_Of_Pictures","Postal_Code","Last_Seen"]
df = df.rename(columns=clean_cols)
df.columns
```

```python
df.describe(include="all")
```

```python
df = df.drop(df.loc[df['Year_Of_Registration'] <= 1990].index, inplace=False)
df["Year_Of_Registration"].value_counts()
```

I have deleted the rows before the year 1990.

```python
df["Brand"].value_counts(normalize=True).sort_values(ascending=False).head(5)
```

Half of the list is consist of German Brand cars. We can deduce that German Cars are more attractive and demanding than other country brands.

```python
mean_price_brand = df.groupby("Brand")["Price"].mean()
round(mean_price_brand).sort_values(ascending=False).head(10)
```

We can observe from the list above that average price due to brand is too high in porche brand than the other brands. The diffence between the closest brand is around 29.000 Euros. It is definitely luxury brand.
