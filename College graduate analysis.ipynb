```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv("recent-grads.csv")
df.head(5)
```

```python
df = df.dropna()
```

# Do students in more popular majors make more money?

```python
df.plot(x="Sample_size" , y="Median" , kind="scatter" , figsize=(12,5) , title = "Major-Money comparison")
```

# How many majors are predominantly male? Predominantly female?

```python
df.plot(x="Major" , y="ShareWomen" , kind="hist" , figsize=(12,5) , title = "Major-Gender comparison")
```

# Which category of majors have the most students?

```python
df.plot.bar(x="Major_category" , y="Total" ,figsize=(12,5) , title = "Major-Student comparison")
```
